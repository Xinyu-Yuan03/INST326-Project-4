#### 2. Online Shopping Cart System
>  Objective: Build a shopping cart system for an online store that manages products, shopping carts, and orders.
>
> Requirements
>>- Use classes to represent products, shopping carts, and orders.
>>- Implement encapsulation to handle product prices and cart contents securely.
>>- Use inheritance to create different types of products (e.g., electronics, clothing).
>>- Demonstrate polymorphism by calculating discounts based on product type.
>>- Include execution code to demonstrate that your solution works

```python
# Solution - enter your code solution below
class Product:
    def __init__(self, name, price):
        self._name = name  # Encapsulated name
        self._price = price  # Encapsulated price

    def get_price(self):
        return self._price

    def __str__(self):
        return f"Product: {self._name}, Price: ${self._price:.2f}"

class Electronics(Product):
    def get_price(self):
        return self._price * 0.9  # 10% discount for electronics

class Clothing(Product):
    def get_price(self):
        return self._price * 0.8  # 20% discount for clothing

class ShoppingCart:
    def __init__(self):
        self._contents = []  # Encapsulated list of products

    def add_product(self, product):
        self._contents.append(product)
        print(f"Added {product} to the cart.")

    def remove_product(self, product):
        if product in self._contents:
            self._contents.remove(product)
            print(f"Removed {product} from the cart.")
        else:
            print(f"{product} not found in the cart.")

    def calculate_total(self):
        return sum(product.get_price() for product in self._contents)

    def list_products(self):
        return "\n".join(str(product) for product in self._contents)

class Order:
    def __init__(self, cart):
        self._cart = cart
        self._total = cart.calculate_total()

    def place_order(self):
        print("Order Summary:")
        print(self._cart.list_products())
        print(f"Total Amount Due: ${self._total:.2f}")

# Execution Code
if __name__ == "__main__":
    # Create products
    laptop = Electronics("Laptop", 2000)
    tshirt = Clothing("T-Shirt", 100)
    headphones = Electronics("Headphones", 200)

    # Create shopping cart
    cart = ShoppingCart()
    cart.add_product(laptop)
    cart.add_product(tshirt)
    cart.add_product(headphones)

    # List products and calculate total
    print("\nCart Contents:")
    print(cart.list_products())
    print(f"Total: ${cart.calculate_total():.2f}\n")

    # Place order
    order = Order(cart)
    order.place_order()
```
